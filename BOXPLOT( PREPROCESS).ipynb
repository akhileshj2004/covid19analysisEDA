# COVID 19 STATEWISE ANALYSIS 

## ->Outlier Detection and Inferences of outliers


### -we shall detect the outliers by using the box plot and scatter plot for all the columns and further go for the outlier treatment 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder  #scikit-learn
from mpl_toolkits.mplot3d import Axes3D

data = pd.read_csv(r"C:\anaconda3\covid_vaccine_statewise.csv")
data
```

```python
# Counting the frequency of each gender
region = data['State'].value_counts()

# Creating the pie chart
plt.figure(figsize=(8, 8))
plt.pie(region, labels=region.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of region')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

# The above pie chart is for catagorical attribute

```python
#COLUMN DATA - Total Doses Administered 

data.boxplot(column ="Total Doses Administered")
plt.title("Boxplot for Total Doses Administered")
plt.show()

m1 = data["Total Doses Administered"].max()
print("The max value in this Column is :")
print(m1)

m2 = data["Total Doses Administered"].min()
print("The min value in this Column is :")
print(m2)

Q1 =  data["Total Doses Administered"].quantile(0.25)
print("The Q1 value in this Column is :")
print(Q1)

Q3 =  data["Total Doses Administered"].quantile(0.75)
print("The Q3 value in this Column is :")
print(Q3)

# Interquartile range(IQR)
IQR = Q3 - Q1
print("The Interquartile range - IQR in this Column is :")
print(IQR)
```
